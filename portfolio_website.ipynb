```python


import gradio as gr

# ‚úÖ Your actual Google Drive File ID
drive_file_id = "11cnFIYkarFVJs8nOL148pUvCYCzz-G6s"

def portfolio_page(page):
    if page == "Home":
        return """
        <h1>üë©‚Äçüíª Hi, I'm Srilekha</h1>
        <p>Welcome to my portfolio website built using Python + Gradio.</p>
        <p>I am a Computer Science graduate passionate about Python, Web Development, and AI.</p>
        """

    elif page == "Projects":
        return """
        <h2>üöÄ My Projects</h2>
        <ul>
            <li><b>DeepFake Detection</b> ‚Äì YOLOv8 + CNN for real-time fake face detection.</li>
            <li><b>Expense Tracker</b> ‚Äì React + Node.js + MongoDB full-stack app.</li>
            <li><b>Portfolio Builder</b> ‚Äì Django-based personal resume/portfolio generator.</li>
            <li><b>Language Translation Tool</b> ‚Äì Python-based translator using Google APIs.</li>
            <li><b>React Quiz App</b> ‚Äì Responsive quiz app using React.js and custom CSS.</li>
        </ul>
        """

    elif page == "Resume":
        return f"""
        <h2>üìÑ Resume</h2>
        <iframe src="https://drive.google.com/file/d/{drive_file_id}/preview"
                width="100%" height="600px"></iframe>
        <p><a href="https://drive.google.com/file/d/{drive_file_id}/view?usp=sharing" target="_blank">
        üì• Download My Resume</a></p>
        """

    elif page == "Contact":
        return """
        <h2>üì¨ Contact Me</h2>
        <p>Email: srilekhabaskar3108@gmail.com</p>
        <p>GitHub: <a href="https://github.com/Srilekha3108" target="_blank">github.com/Srilekha3108</a></p>
        <p>LinkedIn: <a href="https://www.linkedin.com/in/srilekha-baskar-7b44322b1" target="_blank">
        linkedin.com/in/srilekha-baskar-7b44322b1</a></p>
        """

    else:
        return "<h2>Page not found</h2>"

# Build Gradio App
with gr.Blocks(title="üåê Srilekha's Portfolio") as demo:
    gr.Markdown("## üåê Srilekha's Portfolio Website")
    page = gr.Dropdown(["Home", "Projects", "Resume", "Contact"], label="Navigate")
    output = gr.HTML()
    page.change(portfolio_page, inputs=page, outputs=output)

# Launch Website (with public share link)
demo.launch(share=True)
```
